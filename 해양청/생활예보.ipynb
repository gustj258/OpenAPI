```python
from pprint import pprint

import requests
import pandas as pd
import datetime

url = "http://www.khoa.go.kr/oceangrid/grid/api/fcIndexOfType/search.do"

headers = {
    'cache-control': "no-cache",
    'postman-token': "c115584c-ec48-8dff-68b3-6d197189a2cc"
    }

dt_index = pd.date_range(start='20190901', end='20190902')
dt_list = dt_index.strftime("%Y%m%d").tolist()

dat = []
for i in dt_list:
    date=i
    querystring = {"ServiceKey":"OHVMK3dF0Ah6K/oIG71f0g==","Date":date, "ResultType":"json","Type":"SF"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    #print(response.text)
    #print(response.json())
    dat.extend(response.json()["result"]["data"])
    
df = pd.DataFrame(dat)
#df.info()

df["date"] = pd.to_datetime(df["date"])
df["name"] = df["name"].astype(str)
#df["time"] = df["time"].astype('datetime64[ns]')
df["lon"] = df["lon"].astype(float)
df["lat"] = df["lat"].astype(float)
df["air_temp"] = df["air_temp"].astype(str)
#df["weather"] = df["weather"].astype(str)
df["total_score"] = df["total_score"].astype(str)

df.to_excel('predict.xlsx' ,index=False)
```
