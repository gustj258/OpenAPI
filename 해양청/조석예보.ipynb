```python
import requests
import datetime
import pandas as pd


url = "http://www.khoa.go.kr/oceangrid/grid/api/tideObsPreTab/search.do"

headers = {"cache-control": "no-cache", "postman-token": "51b374f3-ba0d-0077-1f3c-ff95b65a81ba"}

dt_index = pd.date_range(start="20190901", end="20200901")
dt_list = dt_index.strftime("%Y%m%d").tolist()
    
data = []
for i in dt_list:
    date = i
    querystring = {
        "ServiceKey": "OHVMK3dF0Ah6K/oIG71f0g==",
        "ObsCode": "DT_0011", # 
        "Date":date,
        "ResultType": "json",
    }
    response = requests.request("GET", url, headers=headers, params=querystring)
    print(response.text)

    data.extend(response.json()["result"]["data"])

df = pd.DataFrame(data)
print(df)
df.to_excel("moring2.xlsx", index=False)
```
