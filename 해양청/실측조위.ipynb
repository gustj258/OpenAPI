```python
from pprint import pprint
import requests
import pandas as pd 
import datetime

url = "http://www.khoa.go.kr/oceangrid/grid/api/tideObs/search.do"

headers = {
    'cache-control': "no-cache",
    'postman-token': "a069457e-601d-e099-1ddd-68acdc071d70"
    }

dt_index = pd.date_range(start='20190901', end='20200901')
dt_list = dt_index.strftime("%Y%m%d").tolist()

#분을 시간으로 출력필요
#고정관측소 = (DT_0001 = 인천) 
dat = []
for i in dt_list:
    date=i
    querystring = {"ServiceKey":"OHVMK3dF0Ah6K/oIG71f0g==","ObsCode":"DT_0061","Date":date,"ResultType":"json"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    #print(response.text)
    #print(response.json())
    dat.extend(response.json()["result"]["data"])
    
df = pd.DataFrame(dat)
#df.info()

df["record_time"] = pd.to_datetime(df["record_time"])
df["tide_level"] = df["tide_level"].astype(float)

#df.info()
##print(df)

df4hours = df.set_index("record_time").resample("4H").mean()

print(df4hours)
df4hours.to_excel('tide30.xlsx')
```
