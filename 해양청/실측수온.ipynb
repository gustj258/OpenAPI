```python
from pprint import pprint

import requests
import pandas as pd 
import datetime

url = "http://www.khoa.go.kr/oceangrid/grid/api/tideObsTemp/search.do"

headers = {
    'cache-control': "no-cache",
    'postman-token': "96839df0-3d9a-c381-b166-2250742a86d8"
    }

dt_index = pd.date_range(start='20200801', end='20200901')
dt_list = dt_index.strftime("%Y%m%d").tolist()

dat = []
for i in dt_list:
    date=i
    querystring = {"ServiceKey":"OHVMK3dF0Ah6K/oIG71f0g==","ObsCode":"DT_0058","Date":date,"ResultType":"json"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    #print(response.text)
    #print(response.json())
    dat.extend(response.json()["result"]["data"])

df = pd.DataFrame(dat)
#df.info()

df["record_time"] = pd.to_datetime(df["record_time"])
df["water_temp"] = df["water_temp"].astype(float)

#df.info()
##print(df)

df4hours = df.set_index("record_time").resample("4H").mean()
#df4hours.info()
print(df4hours)
df4hours.to_excel('temp3.xlsx')

#aUhhR8ZwUV5sBR1RvFpUFA==
```
