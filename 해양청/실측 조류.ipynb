```python
from pprint import pprint
import requests
import pandas as pd 
import datetime

url = "http://www.khoa.go.kr/oceangrid/grid/api/tidalBu/search.do"

headers = {
    'cache-control': "no-cache",
    'postman-token': "82577a9c-5a9a-4677-e954-d6ada392a729"
    }

dt_index = pd.date_range(start='20190901', end='20191201') #날짜 지정하기
dt_list = dt_index.strftime("%Y%m%d").tolist() #strftime 함수를 이용한 년월일로 묶기 

dat = []
for i in dt_list:
    date=i
    querystring = {"ServiceKey":"aUhhR8ZwUV5sBR1RvFpUFA==","ObsCode":"HB_0001","Date":date,"ResultType":"json"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    print(response.text)
    #print(response.josn())
    dat.extend(response.json()["result"]["data"])  #extend 2차원 배열로 합치기
    
df = pd.DataFrame(dat) #pandas 라이브러리를 이용한 데이터 프레임 생성

df["obs_time"] = pd.to_datetime(df["obs_time"]) #각각의 데이터 타입을 지정하기
df["current_spped"] = df["current_speed"].astype(float)
df["current_direct"] = df["current_direct"].astype(float)
                        
#df.info() #데이터 형식 보기
#print(df)
                          
df4hours = df.set_index("obs_time").resample("4H").mean() #4시간 단위로 출력하기
#df4hours.info()
print(df4hours)
df4hours.to_excel('current1.xlsx') #엑셀로 만들기
```
