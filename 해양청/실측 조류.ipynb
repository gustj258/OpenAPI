```python
from pprint import pprint
import requests
import pandas as pd 
import datetime

url = "http://www.khoa.go.kr/oceangrid/grid/api/tidalBu/search.do"

headers = {
    'cache-control': "no-cache",
    'postman-token': "82577a9c-5a9a-4677-e954-d6ada392a729"
    }

dt_index = pd.date_range(start='20190901', end='20191201') # 
dt_list = dt_index.strftime("%Y%m%d").tolist() #strftime     

dat = []
for i in dt_list:
    date=i
    querystring = {"ServiceKey":"aUhhR8ZwUV5sBR1RvFpUFA==","ObsCode":"HB_0001","Date":date,"ResultType":"json"}
    response = requests.request("GET", url, headers=headers, params=querystring)
    print(response.text)
    #print(response.josn())
    dat.extend(response.json()["result"]["data"])  #extend 2  
    
df = pd.DataFrame(dat) #pandas     

df["obs_time"] = pd.to_datetime(df["obs_time"]) #   
df["current_spped"] = df["current_speed"].astype(float)
df["current_direct"] = df["current_direct"].astype(float)
                        
#df.info() #  
#print(df)
                          
df4hours = df.set_index("obs_time").resample("4H").mean() #4  
#df4hours.info()
print(df4hours)
df4hours.to_excel('current1.xlsx') # 
```
